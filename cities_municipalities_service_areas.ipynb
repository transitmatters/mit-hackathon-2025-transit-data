This writeup explains the municipalities in Massachusetts, the City of Boston, and the MBTA Service Areas. It also provides geospatial (GIS) datasets of their boundaries.

We **highly recommend** that your analysis extend to municipalities outside the City of Boston's boundaries (which exclude places like Cambridge), such as to MBTA's Core or Extended Service Areas, or to the entire state. This will be elaborated below.

## Municipalities in Massachusetts and Greater Boston

In Massachusetts, a [municipality](https://gis.data.mass.gov/maps/4b532722f22d43ce9003cb1d118d89cc/explore) is either a **city** (e.g. the City of Boston, the City of Cambridge) or a **town** (e.g. the Town of Brookline). Each municipality has its own government, often with a mayor, a city council, or a town representative.

The term [**"Greater Boston"**](https://en.wikipedia.org/wiki/Greater_Boston) is often used to describe the whole metropolitan region (of varying sizes). This includes, but is not limited to, the City of Boston â€“ which is itself just one of many municipalities in Greater Boston, albeit the most populated one. A more detailed comparison will be shown below.

A common definition of the **"urban"** region of Greater Boston (in contrast to suburban) is [Massachusetts Route 128](https://en.wikipedia.org/wiki/Massachusetts_Route_128), a circumferential highway. Most of Route 128 is concurrent with interstate I-95.

## MBTA Core and Extended Service Areas

The MBTA defines its [**Core Service Area**](https://mbta-massdot.opendata.arcgis.com/datasets/2d7d34175e554ddc9f33c7f0ad6d6872_0/explore) as 65 cities and towns in Massachusetts for which MBTA _bus or rapid transit_ service is provided (including those adjacent to such service). Its [**Extended Service Area**](https://mbta-massdot.opendata.arcgis.com/datasets/ec170028b45e46cea43dd51bda250237_0/explore) is a larger set of 178 cities and towns, which adds those for which _commuter rail_ service is provided (or adjacent).

The MBTA Core Service Area coincides nicely with the "urban" region within Route 128. While the MBTA commuter rail network runs into Rhode Island (particularly Providence), municipalities in Rhode Island are excluded from the MBTA Extended Service Area.

* Municipalities in the [MBTA Extended Service Area](https://mbta-massdot.opendata.arcgis.com/datasets/ec170028b45e46cea43dd51bda250237_0/explore), in indigo

## City of Boston's Neighborhoods

As alluded to above, the City of Boston is one of the many municipalities in the Greater Boston metropolitan area and the MBTA service areas, albeit large in size and the most populous. The [City of Boston's official neighborhoods](https://gis.data.mass.gov/datasets/fefd5fb156d9419e80a3944a62bd64ab_1/explore) contain many _dissolved municipalities_, especially well south of Downtown Boston. They were once municipalities of their own, but were later [annexed](https://en.wikipedia.org/wiki/Municipal_annexation_in_the_United_States) by Boston.

The map below shows all neighborhoods in the City of Boston (in teal), and other municipalities (in red). MBTA rail rapid transit lines are overlaid.

![Map of Boston neighborhoods and other municipalities](images/muni_boston.png)

Observe that many of Boston's neighborhoods (dissolved municipalities), such as Allston and Dorchester, are similar to many nearby municipalities, such as Cambridge and Chelsea, in their size, population and economic connections to Downtown Boston. Meanwhile, many non-Boston municipalities such as Cambridge are important hubs in the downtown core by themselves, and already receive rapid transit service.

Therefore, we **highly recommend** that your analysis extend to municipalities outside the City of Boston's boundaries, such as to MBTA's Core or Extended Service Areas, or to the entire state.
