# MBTA Ridership Data

This page will guide you through datasets that measure MBTA ridership, as well as the [MBTA Rider Census](https://www.mbta.com/performance-metrics/the-rider-census) data with demographic information of public transit riders.

## Visual Exploration of Ridership Data

### TransitMatters Data Dashboard

The [TransitMatters Data Dashboard](https://dashboard.transitmatters.org/) provides weekday ridership data for rapid transit and buses, reported for each week. You can find an overview in the ["Service & Ridership" tab](https://dashboard.transitmatters.org/system/ridership/?startDate=2024-11-12&endDate=2025-11-12):

![Red Line overview, TransitMatters Data Dashboard](images/tm_dashboard_ridership.png)

More granular route-level ridership data for each rapid transit line and each bus route can be found in individual tabs to the left.

The Data Dashboard allows quick exports of query data as CSV files. All ridership data come from MBTA's data sources directly, detailed below.

### MBTA Bus Route Profiles

For ridership on individual bus routes, a good source of visual data is [MBTA Bus Route Profiles](https://www.mbta.com/projects/better-bus-project/update/bus-route-profiles-now-available). Written in 2018, these are detailed reports of each bus route with descriptions and analysis, including scheduled frequencies, ridership, crowding, on-time performance etc. 
Note that the data is pre-Covid.

The bus route profiles illustrate ridership across the length of each route and at various times of the day. For example, here's [bus route 66](https://cdn.mbta.com/sites/default/files/projects/betterbus/route-profiles/66.pdf), one of the most heavily used and most frequent in the network:

![Route 66, weekday inbound ridership by stop](images/bus_profiles_66_stop.png)

![Route 66, weekday ridership by trip](images/bus_profiles_66_trip.png)


## Detailed Ridership Data: MBTA Open Data Portal

The [MBTA Open Data Portal](https://mbta-massdot.opendata.arcgis.com/) is the main source of large-scale ridership data, while also allowing for greater granularity than the TransitMatters Dashboard.

Methodology, example uses, and brief documentations can be found in the [Ridership Guide](https://mbta-massdot.opendata.arcgis.com/documents/c4625d2f71bc4b638ade0e2015851c00/explore). Each data table linked below contains its own documentation.

### Monthly ridership by mode and line

Ridership of each mode (buses, rapid transit, commuter rail, ferry, the RIDE) and each rapid transit line (Blue, Orange, Red, Mattapan, Green, Silver) every month.

* [2018 – present](https://mbta-massdot.opendata.arcgis.com/datasets/2048258a18354256a650d41f8fe4532c_0/explore)
* [2016 – 2018](https://mbta-massdot.opendata.arcgis.com/datasets/a2d15ddd86b34867a31cd4b8e0a83932_0/explore)

### Gated station entries (rapid transit)

These tables report rapid transit ridership as the number of passengers entering each station that has fare gates, primarily using fare validations.

* [Last 2 years](https://mbta-massdot.opendata.arcgis.com/datasets/001c177f07594e7c99f193dde32284c9_0/explore)
* [Historical data by year since 2014](https://mbta-massdot.opendata.arcgis.com/datasets/7859894afb5641ce91a2bb03599fdf5b)

Data is reported for every 30-minute period every day. In particular, this allows you to examine variations across the calendar year.

Notes:
* This excludes passengers boarding the Green Line at surface stops without fare gates. On the other hand, it includes underground stations on the Silver Line (bus rapid transit) that have fare gates.
* At transfer stations, the data only counts riders entering the system there (and then estimates which line they eventually boarded). It excludes riders switching between rapid transit lines at the station (and between SL1/2/3 and the Red Line at South Station), who have no need to go through fare gates.
* These datasets do not distinguish between directions or times of the day.


### On/off ridership by time period (multimodal)

These tables report, for each route in each direction, the number of passengers boarding and alighting at each stop. You can use this to calculate the (average) number of riders onboard vehicles, how ridership is distributed across times of the day and in different directions, etc.

**Rapid transit:**
* By time period: [Fall 2024](https://mbta-massdot.opendata.arcgis.com/datasets/d4610a65064a4d3c8536c75d520e0012_0/explore), [Fall 2023](https://mbta-massdot.opendata.arcgis.com/datasets/80a379ebaa374cfd836ca4d3880ceda4_0/explore), [Fall 2017-19](https://mbta-massdot.opendata.arcgis.com/datasets/9fea7862501f4c59ab2c001ad6ae19b0_0/explore)
* By hour: [Fall 2024](https://mbta-massdot.opendata.arcgis.com/maps/8f20b4e8d24e466fbca157cca63f6e16)

**Buses:**
* By time period: [Fall 2016-24](https://mbta-massdot.opendata.arcgis.com/datasets/7acd353c1a734eb8a23caf46a0e66b23_0/explore), [Spring 2020-21](https://mbta-massdot.opendata.arcgis.com/maps/adbe3a94edcd44259cfee2621ed9f3a7)
* By each scheduled trip: [Fall 2016-24](https://mbta-massdot.opendata.arcgis.com/datasets/8daf4a33925a4df59183f860826d29ee/about)

**Commuter rail:** [By each scheduled trip](https://mbta-massdot.opendata.arcgis.com/datasets/9e8089a985f24bc6a1dbae1e69703808_0/explore) (Spring 2012, Spring 2018 and Fall 2024 only)

**Ferry:** [By each scheduled trip](https://mbta-massdot.opendata.arcgis.com/datasets/ae21643bbe60488db8520cc694f882aa_0/explore)

"Time periods" (e.g. AM peak, midday school) are defined on Page 6 of the [MBTA Service Delivery Policy](https://cdn.mbta.com/sites/default/files/2025-2-7-service-delivery-policy-fall-2024-final.pdf#page=6).

These datasets are averages on a typical weekday, Saturday or Sunday in each year, when the fall (autumn) schedule, and occasionally spring schedule, were in place.

Notes:
* Rapid transit data includes surface stops on the Green Line without fare gates.
* Silver Line (bus rapid transit) is reported under buses.
* At transfer stations, the data counts any rider getting on this line, whether they come from outside the system or a transfer from another rapid transit or bus route.

## Passenger Comfort and Crowding

The [MBTA Service Delivery Policy](https://cdn.mbta.com/sites/default/files/2025-2-7-service-delivery-policy-fall-2024-final.pdf) defines the agency's metrics for **passenger comfort scores**. [This table](https://mbta-massdot.opendata.arcgis.com/datasets/f669fc7a9b1c49b292532b1b7b936e97/explore) contains historical passenger comfort data for each bus and rapid transit route, and for each day type (weekday, Saturday, Sunday).

## Demographics of Riders: MBTA Rider Census

The [**MBTA Rider Census**](https://www.mbta.com/performance-metrics/the-rider-census), or **System-Wide Passenger Survey**, are surveys of riders on all modes (buses, rapid transit, commuter rail). They contain information such as:

* **Demographics** (race, income, etc.)
* **Trip** purpose, frequency, and fares paid
* **Modes of transportation**: accessing the MBTA, transferring within the system, and alternative modes

The following iterations of the Rider Census are available:

* **2024**: The most recent results, with data aggregated from 2022 through 2024 and reported as percentages.
  * [Dataset](https://mbta-massdot.opendata.arcgis.com/datasets/7da1f62034f64cb4bc9e2afefe9a1fdc/explore)
  * [Executive Summary](https://cdn.mbta.com/sites/default/files/2025-07/2025-07-15-rider-census-2024-rider-census-executive-summary-accessible.pdf)
  * [Technical Documentation](https://cdn.mbta.com/sites/default/files/2025-06/2025-06-24-rider-census-2024-rider-census-report-accessible.pdf)
  * An interactive dashboard is available on the [Riders Census page](https://www.mbta.com/performance-metrics/the-rider-census).
  * 2023 and 2022: The survey data from these years are now aggregated into the 2024 datasets above, but each year's executive summary and documentation are still available at the bottom of the [Riders Census page](https://www.mbta.com/performance-metrics/the-rider-census).
* **2015-2017**: A previous iteration that's more comprehensive in some aspects.
  * [This interactive tool](https://www.ctps.org/dv/mbtasurvey2018/index.html) links to the survey results, reports and data downloads.
  * Unique aspects (compared to 2022-24): The 2015-17 survey also provided data in terms of the number of riders, not just percentages. It also has better coverage of transfers between MBTA services.
