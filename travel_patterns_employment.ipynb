# Travel Patterns and Employment Data

This page will guide you through available datasets on travel patterns and demands, as well as on employment data.

## Overview

We present three origin-destination (OD) datasets that describe multimodal travel patterns: employment; Bluebikes trips; and other multimodal travel data. The remainder of this tutorial will mostly focus on employment data.

### 1. Employment: LODES

The US Census Bureau provides official [**LEHD Origin-Destination Employment Statistics (LODES)**](https://lehd.ces.census.gov/data/) data. It describes, for any geographical region based on Census blocks,

* The **Origin-Destination (OD)** data: Where employees working in this region live, and where residents in this region work;
* **Residence Area Characteristics (RAC)**: Demographics of residents living in this region;
* **Workplace Area Characteristics (WAC)**: Demographics of employees working in this region.

### 2. Bike Share Origin-Destination Data: Bluebikes

**Bluebikes**, the leading bikeshare service in the Greater Boston area, publishes comprehensive [**system data**](https://bluebikes.com/system-data) of trip histories, including start and end station and times.

In October 2025, the Boston Region MPO published their study on [BlueBikes and MBTA Connections](https://storymaps.arcgis.com/stories/5cae6946e9914d36a40099aa72eaa5da). Their analysis looks at how people use Bluebikes to connect to MBTA services (or as a replacement of MBTA).

### 3. Multimodal, Multipurpose Travel Demand Data

Companies like StreetLight and Replica estimate real-world travel demand on all modes (including driving, walking and biking) and for all purposes (work, school, social activity, etc), using location-based services. Such data is typically only available for purchase, and thus not included here.

In 2022, MBTA released partial travel demand data from StreetLight [on their website](https://www.mbta.com/projects/bus-network-redesign/update/how-we-used-data-design-equitable-bus-network) as part of their ongoing Bus Network Redesign. The data is available at [**this interactive Tableau dashboard**](https://public.tableau.com/app/profile/mbta.office.of.performance.management.and.innovation/viz/PublicBNRDComparisonDashboard/TravelDemand?publish=yes).

## Employment Data (LODES): Details

### Visual Exploration: OnTheMap

The Census Bureau's [OnTheMap web tool](https://onthemap.ces.census.gov/) displayes LODES data on an interactive map and allows quick queries on any region.

**Polygon queries:** In the "Selection" tab, choose the "Draw Polygon "Freehand" tool, then draw a polygon (double click to finish) and then click "Confirm Selection". A popup will appear with "Perform Analysis on Selection Area", which allows you to specify analysis settings. The following settings and query results show where employees with _private primary_ jobs in the Kendall-MIT area live:

![Query results of where Kendall-MIT workers live](images/onthemap_kendall_workers_live.png)

**Line or point queries:** You can also draw lines and points and add **buffers**, using the "Add Buffer to Selection" section under the "Selection" tab. In the example below, we select areas within 1/2 mile of four arbitrary points:

![Region selection within 1/2 mile of user-defined points](images/onthemap_allston_buffers.png)

OnTheMap also allows importing geographies from KML, SHP and GPS files. Query results can be exported as GIS or tabular data.

### Downloading LODES Data

The [Census Bureau LEHD data page](https://lehd.ces.census.gov/data/) provides methods to download LODES data tables in its own section:

1. Use the [**LED Extraction Tool**](https://ledextract.ces.census.gov/). This is the simplest and most intuitive method, that allows you to specify geographies and job attributes.
2. View and download files directly from the "Download LODES data" section of the LEHD page. Naming conventions are explained in the [LODES Tech Doc](https://lehd.ces.census.gov/doc/help/onthemap/LODESTechDoc.pdf).

The [LODES Tech Doc](https://lehd.ces.census.gov/doc/help/onthemap/LODESTechDoc.pdf) provides documentation for the data. In particular, Page 5 explains the variables for the OD data.

### Job Densities...

Even if you're only interested in non-OD employment data, such as the distribution of all jobs, they can still be obtained from LODES data via the Residence Area Characteristics (RAC) and Workplace Area Characteristics (WAC), or alternatively, by aggregatinv the OD data.

### Joining LODES with Demographic Data: Notes on Blocks vs. Block Groups

The LODES tables downloaded above give data for each Census _block_, which is more granular than the _block groups_ data from [our demographics (US Census) datasets and tutorial](demographics_geospatial.ipynb), namely the Americal Community Survey data.

If you want to join the two geography levels, they can be matched via the **first 12 digits of the geocode**, which identify the block group. Each block's geocode adds 3 digits at the end.
